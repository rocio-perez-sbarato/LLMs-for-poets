# 📚 Parte 1. Introducción teórica a los *Large Language Models*

## ¿Qué es el Procesamiento de Lenguaje Natural?

El Procesamiento de Lenguaje Natural (PLN) es una rama de la Inteligencia Artificial que se enfoca en la interacción entre las computadoras y el lenguaje humano. Su objetivo principal es permitir que las máquinas entiendan, interpreten, generen y manipulen el lenguaje humano de manera significativa. El PLN combina elementos de lingüística, ciencias de la computación e inteligencia artificial para crear sistemas capaces de procesar y analizar grandes cantidades de datos de lenguaje natural.



## ¿Qué es un Modelo de Lenguaje?

Un modelo de lenguaje es un sistema de inteligencia artificial diseñado para entender, procesar y generar lenguaje humano. Estos modelos utilizan técnicas de aprendizaje profundo para capturar patrones y estructuras del lenguaje a partir de grandes cantidades de datos textuales.

### Características principales
1. **Comprensión contextual**: Los modelos de lenguaje modernos pueden entender el contexto y el significado de las palabras en diferentes situaciones.
2. **Generación de texto**: Pueden producir texto coherente y relevante basado en un estímulo inicial.
3. **Versatilidad**: Se pueden aplicar a diversas tareas como traducción, resumen, análisis de sentimientos, entre otras.
4. **Aprendizaje por transferencia**: Los modelos preentrenados pueden adaptarse a tareas específicas con relativamente pocos datos adicionales.

Técnicamente, los modelos de lenguaje ofrecen una forma de asignar una probabilidad a una oración u otra secuencia de palabras, y de predecir una palabra o token a partir de palabras anteriores. ¿Por qué deberíamos preocuparnos por predecir las próximas palabras? La idea de los Modelos de Lenguaje a Gran Escala es que muchas tareas prácticas de Procesamiento del Lenguaje Natural (PNL) se pueden convertir en predicciones de palabras y que un modelo de lenguaje lo suficientemente potente puede resolver con un alto grado de precisión.

> Muchas tareas de PNL (como responder preguntas, resumir, generar sentimientos y traducción automática) se pueden considerar tareas de predicción de palabras y, por lo tanto, se pueden abordar con LLMs.

## ¿Qué es un *Large Language Model*?

La creación de un Modelo de Lenguaje a Gran Escala (LLM, por sus siglas en inglés) normalmente consta de dos pasos:

### 1. Language modeling
El primer paso, llamado preentrenamiento, ocupa la mayor parte del tiempo de cómputo y entrenamiento. Un LLM se entrena en un vasto corpus de texto generalmente extraído de Internet, lo que permite que el modelo aprenda gramática, contexto y patrones del lenguaje. Esta amplia fase de entrenamiento aún no está dirigida a tareas o aplicaciones específicas más allá de predecir la siguiente palabra. El modelo resultante se suele denominar modelo de base. Estos modelos generalmente no siguen instrucciones.

### 2. Fine-tuning
El segundo paso, el ajuste fino o, a veces, el post-entrenamiento, implica utilizar el modelo previamente entrenado y entrenarlo más en una tarea más específica. Esto permite que el LLM se adapte a tareas específicas o exhiba el comportamiento deseado. Por ejemplo, podríamos ajustar un modelo base para que funcione bien en una tarea de clasificación o para seguir instrucciones. Ahorra enormes cantidades de recursos porque la fase de preentrenamiento es bastante costosa y generalmente requiere datos y recursos informáticos que están fuera del alcance de la mayoría de las personas y organizaciones.

## ¿Qué es un Transformer?

- Es una arquitectura de red neuronal que se presentó en el paper ["Attention is All You Need"](https://arxiv.org/abs/1706.03762).
- La idea del transformer es "definir" una palabra de acuerdo a la relación que tiene con las palabras de su vecindario, en una operación de multiplicación matricial con pesos.
- Existen variantes, de acuerdo a que parte de la arquitectura usan:
    - Los modelos de traducción de secuencia a secuencia (e.g. el [Transformer](https://arxiv.org/abs/1706.03762) o el [T5](https://arxiv.org/abs/1910.10683)). Tienen codificador y decodificador. Sirven para tareas de transformación (e.g. traducción).
    - Los modelos basados en el codificador (e.g. [BERT](https://arxiv.org/abs/1810.04805)). Sirven para buscar representaciones vectoriales (embeddings) del texto.
    - Los modelos basados en el decodificador (e.g. [GPT](https://arxiv.org/abs/2005.14165)). Sirven para generación de texto.

## ¿Cómo elegir un modelo que se ajuste a mi tarea?

#### 🌍 Idioma y Dominio

El modelo debe estar entrenado en el idioma y dominio específico de tu tarea. Un modelo entrenado en textos médicos en inglés no funcionará bien para analizar tweets en español.

#### 💻 Tamaño y Recursos

Los modelos más grandes suelen tener mejor rendimiento, pero requieren más recursos computacionales. Considerá tus limitaciones de hardware al elegir.

#### 🎯 Tarea Específica

Algunos modelos están optimizados para tareas particulares como generación de texto, traducción o resumen, por lo que es conveniente elegir uno que se ajuste a tu necesidad específica.

#### 🔧 Fine-tuning vs. Uso Directo

Decidí si necesitás un modelo que puedas usar directamente o si planeás fine-tunearlo con tus propios datos o idioma.

#### 🔍 Interpretabilidad

Para algunas aplicaciones, puede ser crucial entender cómo el modelo llega a sus conclusiones. Algunos modelos ofrecen mejor interpretabilidad que otros.

#### 🧭 Sesgo y Ética

Evaluá si el modelo ha sido auditado por sesgos y considera las implicaciones éticas de su uso en tu aplicación.

> Debido a su capacidad de ser utilizados de tantas maneras, los modelos de lenguaje también tienen el potencial de causar daños. Algunos daños incluyen alucinaciones, prejuicios, estereotipos, desinformación y propaganda, y violaciones de la privacidad y de los derechos de autor.

```python
import os
from huggingface_hub import login
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch
from google.colab import userdata

# Obtener el token del secreto de Colab
hf_token = userdata.get('HUGGINGFACE_TOKEN')

# Realizar el login con el token
if hf_token:
    login(token=hf_token)
else:
    raise ValueError("No se encontró el token de Hugging Face en los secretos de Colab. Por favor, añade el token como un secreto con el nombre 'HUGGINGFACE_TOKEN'.")

def cargar_modelo(nombre_modelo, directorio_local=None, cuantizado=False):
    if directorio_local:
        # Verificar si el modelo ya está descargado
        ruta_local = os.path.join(directorio_local, nombre_modelo)
        if not os.path.exists(ruta_local):
            print(f"Descargando modelo a {ruta_local}...")
            AutoModelForCausalLM.from_pretrained(nombre_modelo).save_pretrained(ruta_local)
            AutoTokenizer.from_pretrained(nombre_modelo).save_pretrained(ruta_local)

        # Cargar desde el directorio local
        modelo = AutoModelForCausalLM.from_pretrained(
            ruta_local,
            device_map="auto",
            torch_dtype=torch.float16,
            quantization_config=BitsAndBytesConfig(load_in_4bit=True) if cuantizado else None
        )
        tokenizador = AutoTokenizer.from_pretrained(ruta_local)
    else:
        # Cargar directamente desde Hugging Face
        modelo = AutoModelForCausalLM.from_pretrained(
            nombre_modelo,
            device_map="auto",
            torch_dtype=torch.float16,
            quantization_config=BitsAndBytesConfig(load_in_4bit=True) if cuantizado else None
        )
        tokenizador = AutoTokenizer.from_pretrained(nombre_modelo)

    return modelo, tokenizador

# Parámetros interactivos
nombre_modelo = "meta-llama/Llama-2-7b-hf" #@param ["meta-llama/Llama-2-7b-hf", "meta-llama/Llama-2-13b-hf", "meta-llama/Llama-2-70b-hf"]
directorio_local = "./modelos_locales" #@param {type:"string"}
cuantizado = True #@param {type:"boolean"}

modelo, tokenizador = cargar_modelo(nombre_modelo, directorio_local, cuantizado)
print(f"\nModelo cargado: {nombre_modelo}")
print(f"Cuantizado: {'Sí' if cuantizado else 'No'}")
print(f"Tamaño del modelo: {modelo.get_memory_footprint() / 1e9:.2f} GB")
```

## ¿Qué cosas podemos hacer con `Transformers`?

La biblioteca Transformers de Hugging Face nos permite realizar una amplia gama de tareas de Procesamiento de Lenguaje Natural de manera sencilla. Veamos algunos ejemplos prácticos:


### ❓Preguntas y Respuestas

```python
from transformers import pipeline

# Cargar el pipeline de preguntas y respuestas
qa_pipeline = pipeline("question-answering",
                       model="distilbert/distilbert-base-cased-distilled-squad")

# Contexto
context = """Para preparar una torta de vainilla, necesitas harina, azúcar, manteca, huevos, esencia de vainilla y polvo de hornear.
Primero, bate la manteca con el azúcar hasta obtener una mezcla cremosa.
Luego, agrega los huevos uno a uno, incorporándolos bien.
Añade la esencia de vainilla y mezcla.
Aparte, tamiza la harina con el polvo de hornear e incorpóralos a la mezcla anterior.
Finalmente, vierte la mezcla en un molde y hornea a 180 grados Celsius durante 30-35 minutos, o hasta que al insertar un palillo salga limpio."""

# Preguntas
questions = [
    "¿En qué paso se añade el azúcar?",
    "¿Cuántos grados necesita el horno?",
    "¿Cuándo se tamiza la harina?"
]

# Responder las preguntas
for question in questions:
    result = qa_pipeline(question=question, context=context)
    print(f"Pregunta: {question}\nRespuesta: {result['answer']}\n")
```

> 😯 Como podemos observar, en la primera pregunta el modelo nos repite el contexto! Esto significa que no puede contestar la pregunta. Las respuestas a las otras preguntas están más cerca, al menos acertando a la oración en donde se encuentra la respuesta.

### 🌎 Traducción de idiomas

```python
from transformers import MarianTokenizer, MarianMTModel
from textwrap import TextWrapper

# Seleccionar el modelo pre-entrenado
nombre_modelo = 'Helsinki-NLP/opus-mt-en-es'

# Obtener el tokenizador y cargar el modelo pre-entrenado
tokenizador = MarianTokenizer.from_pretrained(nombre_modelo)
modelo = MarianMTModel.from_pretrained(nombre_modelo)

def formatear_textos(codigo_idioma, textos):
    return [f">>{codigo_idioma}<< {texto}" for texto in textos]

def realizar_traduccion(textos, modelo, tokenizador, idioma="es"):
    # Preparar los textos en el formato apropiado para el modelo
    textos_formateados = formatear_textos(idioma, textos)

    # Generar traducción usando el modelo
    traducido = modelo.generate(**tokenizador(textos_formateados, return_tensors="pt", padding=True))

    # Convertir los índices de tokens generados de vuelta a texto
    textos_traducidos = [tokenizador.decode(t, skip_special_tokens=True) for t in traducido]

    return textos_traducidos

# Textos de ejemplo en inglés
textos_ingles = [
    "Hello, how are you?",
    "Artificial intelligence is transforming the world.",
    "I love learning new languages."
]

# Realizar la traducción
textos_traducidos = realizar_traduccion(textos_ingles, modelo, tokenizador)

# Crear un wrapper para formatear el texto
wrapper = TextWrapper(width=80)

# Mostrar resultados
for original, traducido in zip(textos_ingles, textos_traducidos):
    print("Texto original:")
    print(wrapper.fill(original))
    print("\nTraducción:")
    print(wrapper.fill(traducido))
    print("\n" + "-"*80 + "\n")
```
